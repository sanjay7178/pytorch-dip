```python
import cv2
import numpy as np
from IPython.display import display, Image

# Load the image
image = cv2.imread('/content/300.jpg')  # Replace with your image path

# 1. Translation
# Define translation matrix
tx, ty = 50, 100  # Shift in x and y directions
translation_matrix = np.float32([[1, 0, tx], [0, 1, ty]])
translated_image = cv2.warpAffine(image, translation_matrix, (image.shape[1], image.shape[0]))

# 2. Rotation
# Define rotation matrix
angle = 45  # Rotation angle in degrees
center = (image.shape[1] // 2, image.shape[0] // 2)  # Center of rotation
rotation_matrix = cv2.getRotationMatrix2D(center, angle, 1.0)  # Scale factor = 1.0
rotated_image = cv2.warpAffine(image, rotation_matrix, (image.shape[1], image.shape[0]))

# 3. Scaling
# Define scaling factor
scale_x, scale_y = 1.5, 0.8  # Scaling factors in x and y directions
scaled_image = cv2.resize(image, None, fx=scale_x, fy=scale_y, interpolation=cv2.INTER_LINEAR)

# Display the transformed images
print("Translation")
display(Image(data=cv2.imencode('.jpg', translated_image)[1].tobytes()))
print("Rotation")
display(Image(data=cv2.imencode('.jpg', rotated_image)[1].tobytes()))
print("Scaling")
display(Image(data=cv2.imencode('.jpg', scaled_image)[1].tobytes()))
```
